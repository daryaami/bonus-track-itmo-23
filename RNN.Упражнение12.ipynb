```python
import warnings
warnings.filterwarnings('ignore')

import gensim
from gensim.models import KeyedVectors
```

```python
import numpy as np
import pandas as pd
```

Структура называется <code>KeyedVectors</code> и по сути представляет собой отображение между ключами и векторами. Каждый вектор идентифицируется своим ключом поиска, чаще всего коротким строковым токеном, поэтому обычно это соответствие между



```python
# Загрузка модели google news vectors
#  = KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300.bin", binary=True)
gn_model = KeyedVectors.load_word2vec_format("/content/googlenewsvectorsnegative300/GoogleNews-vectors-negative300.bin", 
                                      binary=True, 
                                      encoding='utf8')

# Загрузка модели на основе британского национального корпуса
br_model = KeyedVectors.load_word2vec_format("model.bin", binary=True)
```

```python
# Определение косинусного расстояния 
def cosine_distance(w1, w2, model):
  return(model.distance(w1, w2))

two_words = ['cucumber', 'doctor']
two_words_br = [i+'_NOUN' for i in two_words]
```

```python
# Вывод результатов
print("Результат модели GN:", round(cosine_distance(*two_words, gn_model), 3))
print("Результат модели BR:", round(cosine_distance(*two_words_br, br_model), 3))
```

```python
words = ['cucumber', 'doctor', 'professor', 'student']
words_br = [i+'_NOUN' for i in words]
```

```python
word_similarity = {word : [] for word in words}
for i in words:
  for j in words:
    if i != j:
      word_similarity[i].append(cosine_distance(i, j, gn_model))
  print(i, np.mean(word_similarity[i]))
```

```python
word_similarity_br = {word : [] for word in words_br}
for i in words_br:
  for j in words_br:
    if i != j:
      word_similarity_br[i].append(cosine_distance(i, j, br_model))
  print(i, np.mean(word_similarity_br[i]))
```

```python
import pandas as pd

df = pd.read_csv("wordsim353_sim_rel/wordsim_similarity_goldstandard.txt", 
                 sep="\t", header=None)
df.columns = ["first", "second", "score"]
df = df.iloc[16:116]
df
```

```python
gn_dist, br_dist, scores = [], [], []
gn_dist_2, br_dist_2 = [], []

for row in df.iterrows():
    #enter your code here
    
  w1, w2 = row[1]["first"], row[1]["second"]
  try:
    gn = gn_model.similarity(w1, w2)
    br = br_model.similarity(w1.lower() + "_NOUN", w2.lower() + "_NOUN")
    gn_dist.append(gn)
    br_dist.append(br)
    br_dist_2.append(1 - br_model.distance(w1.lower() + "_NOUN", w2.lower() + "_NOUN"))
    gn_dist_2.append(1 - gn_model.distance(w1, w2))
    scores.append(row[1]["score"])
  except KeyError as e:
    print(e, "Skipping this word.")
```

```python
from scipy.stats import spearmanr

#enter your code here

print("GN spearmanr corr:", spearmanr(scores, gn_dist).correlation)
print("British spearmanr corr:", spearmanr(scores, br_dist).correlation)
```
